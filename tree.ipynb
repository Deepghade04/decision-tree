# **Target and Splitting Data**

```python
y=iris.target
```

```python
y
```

```python
#add target column to dataframe
df['target'] = y
df['target_names'] = df['target'].apply(lambda x: iris.target_names[x])
```

```python
from sklearn.model_selection import train_test_split
x = df[iris.feature_names]  # Features
y = df['target']            # Target
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
modell = DecisionTreeClassifier(random_state=42)
modell = DecisionTreeClassifier()
```

```python
modell.fit(x_train, y_train)
```

```python
modell.score(x_train,y_train)
```

```python
modell.score(x_test,y_test)
```

```python
from sklearn import tree
plt.figure(figsize=(12,8))
tree.plot_tree(modell, feature_names=iris.feature_names, class_names=iris.target_names, filled=True)
plt.show()
```

# **Prediction**

```python
new_data = np.array([[5.1, 3.5, 1.4, 0.2], [6.7, 3.1, 4.7, 1.5]])
```

```python
predictions = modell.predict(new_data)
predicted_classes = iris.target_names[predictions]

print("Predictions:", predictions)
print("Predicted Classes:", predicted_classes)
```
